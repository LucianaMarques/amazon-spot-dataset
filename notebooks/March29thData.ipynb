# AWS Spot instances price data visualization

We here present data on spot instances' price variation over a three month period for us-east-1 availability zone.

```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

## Dataset preparation 

Preparing the dataset

### t2.micro machines

```python
with open('t2micro.json') as json_file:
    price_data = json.load(json_file)['SpotPriceHistory']
    price_dict = {}
    price_dict['ProductDescription'] = []
    price_dict['SpotPrice'] = []
    price_dict['Timestamp'] = []
    for data in price_data:
        price_dict['ProductDescription'].append(data['ProductDescription'])
        price_dict['SpotPrice'].append(float(data['SpotPrice']))
        price_dict['Timestamp'].append(data['Timestamp'])    
    
df_t2micro = pd.DataFrame(data=price_dict)
df_t2micro = df_t2micro.set_index('Timestamp')
```

```python
windows_data = df_t2micro[df_t2micro['ProductDescription']=='Windows'].drop(columns=['ProductDescription'])
plt.figure()
windows_data.plot()
```

```python
linux_data = df_t2micro[df_t2micro['ProductDescription']=='Linux/UNIX'].drop(columns=['ProductDescription'])
plt.figure()
linux_data.plot()
```

```python
suse_data = df_t2micro[df_t2micro['ProductDescription']=='SUSE Linux'].drop(columns=['ProductDescription'])
plt.figure()
suse_data.plot()
```

```python
suse_data = df_t2micro[df_t2micro['ProductDescription']=='Red Hat Enterprise Linux'].drop(columns=['ProductDescription'])
plt.figure()
suse_data.plot()
```
